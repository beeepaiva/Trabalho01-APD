```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import math
%matplotlib inline
plt.style.use('seaborn-whitegrid')

dataSeq = pd.read_csv('exemplo-Sequencial.csv')
dataOpenMP = pd.read_csv('exemplo-OpenMP.csv')

x = [dataSeq.columns[0], dataSeq.columns[1], dataSeq.columns[2], dataSeq.columns[3]]

y_Seq = [np.mean(dataSeq.iloc[:,0]), 
     np.mean(dataSeq.iloc[:,1]), 
     np.mean(dataSeq.iloc[:,2]), 
     np.mean(dataSeq.iloc[:,3])]

y_OMP = [np.mean(dataOpenMP.iloc[:,0]), 
     np.mean(dataOpenMP.iloc[:,1]), 
     np.mean(dataOpenMP.iloc[:,2]), 
     np.mean(dataOpenMP.iloc[:,3])]

std = [dataOpenMP.iloc[:,0].std(), 
       dataOpenMP.iloc[:,1].std(), 
       dataOpenMP.iloc[:,2].std(), 
       dataOpenMP.iloc[:,3].std()]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.errorbar(x, y_OMP, yerr=std, fmt='.k');
ax.scatter(x,y_Seq, marker="^", color="black")
ax.scatter(x,y_OMP, marker="x", color="black")
ax.set_yscale('log')
ax.set_title("Processamento")
ax.set_ylabel('Tempo de execução em segundos')
ax.set_xlabel('Imagem em pixels')
```
